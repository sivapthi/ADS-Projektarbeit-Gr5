```python
import http.client

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

conn = http.client.HTTPSConnection("www.openligadb.de")

headers = {
    'Connection': 'keep-alive'
    }


conn.request("GET", "/api/getmatchdata/PL/2016", headers=headers)


res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))

# Keine vollstndigen Daten erhltlich auf der Homepage von Open Liga fr Saison 2017/16
conn.request("GET", "/api/getmatchdata/PL/2017", headers=headers)

res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))

conn.request("GET", "/api/getmatchdata/enpl/2018", headers=headers)

res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))

#Keine Daten erhltlich auf der Homepage von Open Liga fr Saison 2020 - 2021
conn.request("GET", "/api/getmatchdata/PL/2019", headers=headers)

res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))


#Keine Daten erhltlich auf der Homepage von Open Liga fr Saison 2020 - 2021
conn.request("GET", "/api/getmatchdata/PL/2020", headers=headers)

res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))
```
